````python
from sklearn.svm import SVC, LinearSVC
from sklearn.linear_model import SGDClassifier


columnas_numericas_standard = ['Height', 'Weight', 'Age','Aggression', 'Reactions',
                               'Composure', 'Long_Pass', 'Acceleration','Speed', 'Stamina',
                               'Strength', 'Balance', 'Agility','Jumping', 'Penalties']

columnas_numericas_minmax = ['Ball_Control', 'Dribbling', 'Marking','Sliding_Tackle','Standing_Tackle','Interceptions','Vision',
                             'Crossing', 'Short_Pass', 'Heading','Shot_Power','Finishing','Long_Shots','Curve',
                             'Freekick_Accuracy', 'Volleys']
columna_categorica = ['Nationality', 'Club_Position','Name','Preffered_Foot',
                      'Work_Rate', 'Weak_foot','Skill_Moves']

# Definir los transformadores para cada tipo de columna
transformadores = [
    ('escala_minmax', MinMaxScaler(), columnas_numericas_minmax),
    ('escala_estandar', StandardScaler(), columnas_numericas_standard),
    ('one-hot', OneHotEncoder(handle_unknown='ignore'), columna_categorica)
]

# Crear el ColumnTransformer
preprocessor = ColumnTransformer(transformadores)


# Definir los clasificadores
classifier1 = LinearSVC()  # SVC lineal
classifier2 = SVC()  # SVC
classifier3 = SGDClassifier()  # SGD classifier

# Definir los pipelines utilizando el ColumnTransformer definido anteriormente
pipeline1 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier1)
])

pipeline2 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier2)
])

pipeline3 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier3)
])
````

Para lograr esto, primero, les pide que etiqueten de la siguiente manera los valores que aparecen en el atributo ```Club_Position```, pidiendo que agrupen los valores en los siguientes grupos:

````python
etiquetas = {
    'ataque': ['ST', 'CF'],
    'central_ataque': ['RW', 'CAM', 'LW'],
    'central': ['RM', 'CM', 'LM'],
    'central_defensa': ['RWB', 'CDM', 'LWB'],
    'defensa': ['RB', 'CB', 'LB'],
    'arquero': ['GK']
}

# Crear una nueva columna con las etiquetas correspondientes
df['Position_Label'] = df['Club_Position'] \
.apply(lambda x: next((etiqueta for etiqueta, valores in etiquetas.items() if x in valores), None))

df.dropna(subset=['Position_Label'], inplace=True)

# Contar cu√°ntos ejemplos hay por clase
conteo_clases = df['Position_Label'].value_counts()
print(conteo_clases)
````

````python
columna_categorica = ['Nationality','Name','National_Position','Preffered_Foot',
                      'Work_Rate', 'Weak_foot','Skill_Moves']

# Definir los transformadores para cada tipo de columna
transformadores = [
    ('escala_minmax', MinMaxScaler(), columnas_numericas_minmax),
    ('escala_estandar', StandardScaler(), columnas_numericas_standard),
    ('one-hot', OneHotEncoder(handle_unknown='ignore'), columna_categorica)
]

# Crear el ColumnTransformer
preprocessor = ColumnTransformer(transformadores)


# Definir los clasificadores
classifier1 = LinearSVC()  # SVC lineal
classifier2 = SVC()  # SVC
classifier3 = SGDClassifier()  # SGD classifier

# Definir los pipelines utilizando el ColumnTransformer definido anteriormente
pipeline1 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier1)
])

pipeline2 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier2)
])

pipeline3 = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier3)
])
````

````python
X = df.drop(['Label','Position_Label','Club_Position'], axis = 1)
y = df['Position_Label']

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

pipeline1.fit(X_train, y_train)
pipeline2.fit(X_train, y_train)
pipeline3.fit(X_train, y_train)

y_pred1 = pipeline1.predict(X_test)
y_pred2 = pipeline2.predict(X_test)
y_pred3 = pipeline3.predict(X_test)

# Obtener el informe de clasificaci√≥n
print("Pipeline 1:")
print(classification_report(y_test, y_pred1))

print("Pipeline 2:")
print(classification_report(y_test, y_pred2))

print("Pipeline 3:")
print(classification_report(y_test, y_pred3))
````
