La manera en la que se presenta un commit al usar este comando es información como el hash asociado al commit, fecha y hora, autor, el mensaje creado al momento de hacer el commit y detalles de los principales cambios realizados en el commit.

6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.

6. Falso. Puede pasar que dos personas se encuentren trabajando en el mismo codigo fuente pero en ramas distintas, al intentar el comando merge en estas dos ramas, Git notificará que existe un conflicto ya que existen versiones distintas que ocasionaran problemas al fusionar estas ramas. Git no resolverá este error automaticamente sino que se tendrá que modificar manualmente el código antes de hacer el merge.

7. Falso. El comando git pull se utiliza para actualizar los cambios de un reposito remoto en el repositorio local, git clone por el contrario se utiliza para clonar un repositorio remoto creandose una copia de este a nivel local. 


8. Falso. Cada persona cuenta con una copia local del repositorio. Esto crea copias individuales en los que cada miembro podrá hacer cambios sin afectar las copias del resto del equipo. 9. Verdadero. Se puede utilizar git reset, git revert o git checkout para deshacer los cambios y volver a una versión anterior del repositorio. La razón de poder hacer esto es que al ser un DVCS, Git almacena los cambios realizados en el historial del repositorio, lo que permite acceder a distintas versiones del archivo o código, incluyendo cambios que hayan sido eliminados por error.

10. Falso. El comando git commit se utiliza para registrar los cambios en el repositorio local. Para enviar los cambios al repositorio remoto, se usa git push.
